### Creating Data

```python
#creating series from lists
names = ['Rexy', 'Raphael','Rita','Razel','Randiel']
weight= [70, 75, 78, 73, 72]
names
```

```python
SeriesNames = pd.Series(names, index = weight )
SeriesNames
```

```python
SeriesNames[70]
```

```python
SeriesNames2= pd.Series(['Damalie','Damon','Durim','Donald'], index = ['A','B','C','D'])
SeriesNames2
```

```python
SeriesNames2['D']
```

```python
#Creating Series from dictionaries
city={0:'Kampala',1:'Arusha',2:'Nyeri',3:'Addis',4:'Accra'}
```

```python
cities = pd.Series(city)
cities
```

```python
#creating a dataframe
db=pd.DataFrame(columns='City Size Population'.split())
db['City']='Kampala Arusha Nyeri Addis Accra'.split()
db['Size']=1000,5000,24000,4300,6000
db['Population']=250000,150000,230000,500000,100000
db
```

```python
#Querying the data to get size, population
db[['Size','Population']]
```

```python
#Adding a new column called Population desnity from population & Size
db ['Population Density']= db['Population']/db['Size']
db
```

### Dealing with already existing Data

```python
covid_data = pd.read_csv('covidset.csv')
covid_data
```

```python
covid_data.columns
```

#### Selection and Indexing

```python
covid_data[['cases','deaths']]
```

*Data Frame columns are just series*

#### Selecting Rows

```python
covid_data.iloc[0]
```

```python
covid_data.iloc[2]
```

#### Selecting subset of rows and columns

```python
covid_data.loc[0,'cases']
```

```python
covid_data.loc[[0,2,4],['cases','deaths']]
```

#### Conditional Selection

An important feature of pandas is conditional selection using bracket notation, very similar to numpy:

```python
covid_data.columns
covid_data[(covid_data['month']==12)&(covid_data['year']==2019)]
```

```python
#More than one condition
(covid_data.loc[5475,covid_data.columns])
```

#### Groupby

The groupby method allows you to group rows of data together and call aggregate functions

```python
continent_Exp = covid_data.groupby('continentExp')
continent_Exp
```

And then call aggregate methods off the object:

```python
last = ['cases','deaths']
continent_Exp[last].mean()
```

```python
continent_Exp[last].std()
```

```python
continent_Exp[last].max()
```

```python
continent_Exp[last].sum()
```

```python
continent_Exp[last].count()
```

```python
continent_Exp['cases'].count()['Africa']
```

```python
continent_Exp[last].describe()
```

```python
continent_Exp[last].describe().transpose()
```

```python
continent_Exp['cases'].describe().transpose()
```

```python
#Number of countries reported on in the data
covid_data['countriesAndTerritories'].nunique()
```

```python
#Specify the countries captured in the data
c = covid_data['countriesAndTerritories'].unique()
print('Countries are:\n',c)
```

```python
#Number of continents in the data
print('Number of Continents: ',covid_data['continentExp'].nunique())
```

```python
#Which country (ies) is(are) included in the continent named 'Other' in the data?
df = covid_data[(covid_data['continentExp']=='Other')]
print('The country(ies) are:',df['countriesAndTerritories'].unique())
```

```python
#How many African countries are in the data?
dt = covid_data[(covid_data['continentExp']=='Africa')]
print('There are',dt['countriesAndTerritories'].nunique() ,'countries in Africa')
```

```python
#Which African country had the highest total cases in the month of June?
sum_2 = dt[dt['month'] == 6] 
print(sum_2.groupby('countriesAndTerritories')['cases'].sum().idxmax())
```

```python
#When did Uganda register the first covid case?
fff = covid_data[(covid_data['countriesAndTerritories'] == 'Uganda')].reset_index()
fff.drop('index',axis= 1,inplace = True)
```

```python
fff.head(92)
```

```python
fff['dateRep_new'] = pd.to_datetime(fff['dateRep'])
print(fff.groupby('countriesAndTerritories')['dateRep'].min()['Uganda'])
```

```python
#Question 10
#How many countries have zero deaths uptodate?
death_ = covid_data.groupby('countriesAndTerritories')
death_['deaths'].sum().value_counts().iloc[0]
```
